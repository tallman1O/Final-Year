# ZERO-SHOT CLASSIFICATION

The Below pipline is called zero-shot because you don't need to fine-tune the model on your data to use it.
It can directly return probablitiy scores for any list of labels you want.

```python
classifier = pipeline("zero-shot-classification")
candidate_labels = ["education", "politics", "business"]
classifier(
    "This is a course about the Transformers library, and HuggingFace OMG",
    candidate_labels=candidate_labels,
)
```

```python
candidate_labels = ["entertainment", "sports", "education"]
classifier(
    "This is a course about the Transformers library, and Lawn Tennis",
    candidate_labels=candidate_labels,
)
```

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models, which can be used to build AI Chat Applications.", top_k=2)
```

```python
ner = pipeline("ner", aggregation_strategy="simple")
ner("My name is Mehul and I work at Rakuten in Tokyo")
```

```python
qna = pipeline("question-answering")
qna(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```
